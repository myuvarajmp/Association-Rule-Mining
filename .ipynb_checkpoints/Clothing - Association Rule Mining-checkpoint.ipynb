```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Loading the data
basket = pd.read_csv('customer_baskets.csv')
```

```python
basket.head()
```

```python
#Converting the transactions into row format
new_basket = pd.DataFrame(basket.basket.str.split(',').tolist(), index=basket.customer_id).stack()
new_basket.to_frame()
new_basket = new_basket.reset_index([0, 'customer_id'])
new_basket.columns = ['customer_id', 'basket']
```

```python
new_basket.basket=new_basket.basket.astype(int)
```

```python
category = pd.read_csv('clothing_categories.csv')
```

```python
new_basket = pd.merge(left=category,right=new_basket,left_on='category_id',right_on='basket')
```

```python
new_basket = new_basket.loc[:,['name','customer_id']]
```

```python
new_basket.sort_values(by='customer_id')
```

```python
basket_pivot = new_basket.pivot_table(index='customer_id',columns='name',values='name',aggfunc={'name':'count'})
```

```python
basket_pivot.head()
```

```python
basket_pivot.shape
```

```python
#Filling NaN with '0'
new_basket = basket_pivot.fillna(0)
```

```python
new_basket.head()
```

```python
#Converting all transactions have item count greater than 1 to 1 for running apriori algorithm
for i in new_basket.columns:
    new_basket.loc[new_basket[i] > 0, i] = 1
```

```python
new_basket.head()
```

## Importing Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Since the store has only 48 items in it, minimum support threshold is set as 0.05.
#An itemset to be present in 5% of the total transactions
freq_itemset = apriori(df = new_basket, min_support = 0.05, use_colnames = True)
```

```python
rules = association_rules(freq_itemset,metric="confidence", min_threshold = 0.5)
```

### Sorting values by confidence and lift

```python
#Top 10 itemset in our transaction
rules.sort_values(by = ['confidence','lift'],ascending = False).head(10)
```

### Since "Ball Gown" is very common in all rules, so exempting them from consequents to find other important associations

```python
s1 = frozenset({'Ball Gown'})
x = pd.DataFrame()
for index, row in rules.iterrows():
    if (row["consequents"].isdisjoint(s1)):
        x = x.append(pd.DataFrame([list(row)], columns =rules.columns))
x.reset_index(drop=True, inplace=True)
```

```python
x.sort_values(by = ['lift','confidence'], ascending = False).head(50)
```
